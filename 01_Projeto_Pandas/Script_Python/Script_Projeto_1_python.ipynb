##### Visualizando as informações

```python
#Quantas fumantes?
df.groupby("FUMA?")["id"].count()
```

```python
#Quantas com DSTs
df.groupby("DSTs")["id"].count()
```

```python
#Maior e média de tempo utlizando o DIU
print(f'{df["DIU_ANOS"].max()} e {df["DIU_ANOS"].mean():.0}')
#Maior e média de tempo de fumantes
print(f'{df["TEMPO_FUMA(ANOS)"].max()} e {df["TEMPO_FUMA(ANOS)"].mean():.2}')
#Maior e média de tempo utilizando anti-concepcional
print(f'{df["ANTI_CONCEPCIONAL_ANOS"].max()} e {df["ANTI_CONCEPCIONAL_ANOS"].mean():.2}')
```

```python
#Idade mais nova e idade mais velha
print(f'Mulher mais velha {df["IDADE"].max()} e mulher mais nova {df["IDADE"].min()}')
```

```python
#Visualizando mulheres por idade
df.groupby(df["IDADE"])["id"].count().plot.barh()
plt.title("Mulheres por IDADE", fontweight="bold")
plt.xticks(rotation = "horizontal")
plt.yticks([5, 25, 40])
plt.xlabel("Quantidade")
plt.savefig("Mulheres por IDADE.png", dpi = 300, bbox_inches='tight');
```

```python
#Visualizando quantidade de DSTs por idade/DST
plt.hist(
    df.groupby(df["IDADE"])[
        [
            "condilomatose",
            "cervical_condilomatose",
            "condilomatose_vaginal",
            "vulvoperineal_condilomatose",
            "sífilis",
            "doença_inflamatória_pélvica",
            "herpes_genital",
            "molusco_contagioso",
            "AIDS",
            "HIV",
            "hepatite_b",
            "HPV",
        ]
    ].sum(),
    bins=5,
)
plt.title("Quantidade de DSTs por IDADE", fontweight="bold")
plt.ylabel("IDADE")
plt.xlabel("Quantidade")
plt.xticks([1,2,3,4,5])
plt.legend(
    df[
        [
            "condilomatose",
            "cervical_condilomatose",
            "condilomatose_vaginal",
            "vulvoperineal_condilomatose",
            "sífilis",
            "doença_inflamatória_pélvica",
            "herpes_genital",
            "molusco_contagioso",
            "AIDS",
            "HIV",
            "hepatite_b",
            "HPV",
        ]
    ],
    loc="center",
    bbox_to_anchor=(0.5, 1.4),
    ncol=2,
    fancybox=True,
    shadow=True,
)
plt.savefig("QTD_DSTs por IDADE.png", dpi = 300, bbox_inches='tight');
```

```python
#Visualizando quais idades fizeram EXAME DX
plt.scatter(x=df["EXM_DX"], y=df["IDADE"])
plt.title("Exame DX por IDADE", fontweight="bold")
plt.ylabel("IDADE")
plt.xlabel("Quantidade")
plt.savefig("Exame DX por IDADE", dpi = 300, bbox_inches='tight');
```

```python
#Visualizando em que idade aparece mais o HPV
df.groupby("IDADE")["HPV"].sum().plot()
plt.title("hpv por IDADE", fontweight="bold")
plt.xlabel("IDADE")
plt.ylabel("Quantidade")
plt.savefig("Idades recorrentes de HPV", dpi = 300, bbox_inches='tight');
```

```python
#Visualizando quantidade de fumantes e não fumantes no dataset
df.groupby("FUMA?")["id"].count().plot.bar()
plt.xticks(ticks= [0, 1], labels=["Não Fumantes", "Fumantes"], rotation = "horizontal")
plt.title("Quantidade de Fumantes", fontweight="bold")
plt.xlabel("TIPO")
plt.ylabel("Quantidade")
plt.savefig("QTD_fumantes.png", dpi = 300, bbox_inches='tight');
```

```python
#Filtrando apenas as fumantes
fumantes = df[df["FUMA?"] == True]
```

```python
#Visualizando EXAMES DX por idade dos fumantes
fumantes.groupby("IDADE")[["EXM_DX_CANCER", "EXM_DX_CIN", "EXM_DX_HPV", "EXM_DX"]].sum().plot()
plt.xlabel("Idade")
plt.xticks(rotation = "horizontal")
plt.ylabel("Quantidade")
plt.yticks([2])
plt.legend()
plt.savefig("FUMANTES_QTD_EXM_DX_P_IDADE.png", dpi = 300, bbox_inches='tight');
```

```python
#Visualizando Parceiros Sexuais e pessoas com DSTs
df.groupby("QTD_PARC_SEXUAIS")[["condilomatose",
            "cervical_condilomatose",
            "condilomatose_vaginal",
            "vulvoperineal_condilomatose",
            "sífilis",
            "doença_inflamatória_pélvica",
            "herpes_genital",
            "molusco_contagioso",
            "AIDS",
            "HIV",
            "hepatite_b",
            "HPV"]].sum().plot()
plt.title("Quantidade de DSTs por qtd. de parceiros Sexuais", fontweight="bold")
plt.ylabel("Nº de Pessoas com DSTs")
plt.xlabel("Qtd. de Parceiros Sexuais")
plt.legend(
    df[
        [
            "condilomatose",
            "cervical_condilomatose",
            "condilomatose_vaginal",
            "vulvoperineal_condilomatose",
            "sífilis",
            "doença_inflamatória_pélvica",
            "herpes_genital",
            "molusco_contagioso",
            "AIDS",
            "HIV",
            "hepatite_b",
            "HPV",
        ]
    ],
    loc="center",
    bbox_to_anchor=(0.5, 1.4),
    ncol=2,
    fancybox=True,
    shadow=True,
)
plt.savefig("DSTs por Quantidade de parceiros sexuais.png", dpi = 300, bbox_inches='tight');
```
